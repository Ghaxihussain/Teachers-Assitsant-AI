```python
from dotenv import load_dotenv
import os
import boto3

load_dotenv()  # ← first

# then create client AFTER loading env
s3 = boto3.client(
    "s3",
    aws_access_key_id=os.getenv("AWS_ACCESS_KEY_ID1"),
    aws_secret_access_key=os.getenv("AWS_SECRET_ACCESS_KEY1"),
    region_name="us-east-1"  # hardcode for now since REGION is None
)

response = s3.list_objects_v2(Bucket="ghazihussainnbucket1")
```

```python
def store_s3(file, name):
    s3.upload_file(file, "ghazihussainnbucket1", "rag_uploads/{name}.pdf")
```

```python
from fastapi import FastAPI, UploadFile
 

app = FastAPI()


@app.post("up")
def upload(file :UploadFile):
    store_s3(file, "file1")
```
